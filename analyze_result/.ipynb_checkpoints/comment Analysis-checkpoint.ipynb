```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Given your odds ratio results
odds_ratio_results = pd.DataFrame({
    'Dimension': ['knowledge', 'identity', 'power', 'conflict', 'similarity', 
                 'fun', 'status', 'trust', 'support', 'romance'],
    'Odds_Ratio': [2.888, 0.575, 0.452, 0.426, 0.388, 
                   0.383, 0.000, 0.000, 0.000, 0.000],
    'P_Value': [0.000, 0.003, 0.087, 0.000, 0.030, 
                0.257, 0.184, 0.000, 0.006, 1.000],
    'CI_Lower': [2.294, 0.393, 0.187, 0.316, 0.161, 
                 0.095, 0.013, 0.004, 0.005, 0.043],
    'CI_Upper': [3.636, 0.843, 1.093, 0.576, 0.938, 
                 1.540, 3.252, 0.902, 1.307, 11.293],
    'Count': [29171, 3955, 922, 8307, 1069, 
              437, 206, 732, 508, 60]
})

# Create the plot
plt.figure(figsize=(12, 6))

# Plot the odds ratios as points
plt.scatter(range(len(odds_ratio_results)), odds_ratio_results['Odds_Ratio'], 
           color='blue', s=100, zorder=5)

# Add error bars
for i in range(len(odds_ratio_results)):
    plt.vlines(i, 
               odds_ratio_results['CI_Lower'].iloc[i],
               odds_ratio_results['CI_Upper'].iloc[i],
               color='blue', alpha=0.5)

# Add horizontal line at OR = 1
plt.axhline(y=1, color='r', linestyle='--', alpha=0.5, label='No Effect Line')

# Customize plot
plt.xticks(range(len(odds_ratio_results)), 
           odds_ratio_results['Dimension'],
           rotation=45,
           ha='right')
plt.ylabel('Odds Ratio (log scale)')
plt.xlabel('Dimension')
plt.title('Odds Ratios by Dimension (with 95% CI)')
plt.yscale('log')

# Add significance indicators
for i, p in enumerate(odds_ratio_results['P_Value']):
    y_pos = odds_ratio_results['CI_Upper'].iloc[i]
    if p < 0.05:
        plt.text(i, y_pos*1.1, '*', ha='center', va='bottom', fontsize=12)
    if p < 0.01:
        plt.text(i, y_pos*1.1, '**', ha='center', va='bottom', fontsize=12)
    if p < 0.001:
        plt.text(i, y_pos*1.1, '***', ha='center', va='bottom', fontsize=12)

# Add legend for significance
plt.text(1.02, 0.98, '* p<0.05\n** p<0.01\n*** p<0.001', 
         transform=plt.gca().transAxes, 
         verticalalignment='top')

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Print interpretation
print("\nInterpretation of Odds Ratios:")
print("\nSignificant positive associations:")
for _, row in odds_ratio_results[
    (odds_ratio_results['Odds_Ratio'] > 1) & 
    (odds_ratio_results['P_Value'] < 0.05)
].iterrows():
    print(f"\n{row['Dimension']}:")
    print(f"- {row['Odds_Ratio']:.2f} times more likely to receive delta")
    print(f"- 95% CI: [{row['CI_Lower']:.2f}, {row['CI_Upper']:.2f}]")
    print(f"- P-value: {row['P_Value']:.4f}")

print("\nSignificant negative associations:")
for _, row in odds_ratio_results[
    (odds_ratio_results['Odds_Ratio'] < 1) & 
    (odds_ratio_results['P_Value'] < 0.05)
].iterrows():
    print(f"\n{row['Dimension']}:")
    print(f"- {row['Odds_Ratio']:.2f} times less likely to receive delta")
    print(f"- 95% CI: [{row['CI_Lower']:.2f}, {row['CI_Upper']:.2f}]")
    print(f"- P-value: {row['P_Value']:.4f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import fisher_exact

def calculate_odds_ratios_for_dimension(data, dimension_column='dimension'):
    dimensions = []
    odds_ratios = []
    p_values = []
    ci_lowers = []
    ci_uppers = []
    total_counts = []
    delta_counts = []
    
    for dim in data[dimension_column].unique():
        # Create contingency table
        contingency = pd.crosstab(data[dimension_column] == dim, data['has_delta'])
        
        # Calculate odds ratio
        odds_ratio, p_value = fisher_exact(contingency)
        
        # Calculate confidence intervals
        total = len(data[data[dimension_column] == dim])
        success = len(data[(data[dimension_column] == dim) & (data['has_delta'] == 1)])
        
        dimensions.append(dim)
        odds_ratios.append(odds_ratio)
        p_values.append(p_value)
        total_counts.append(total)
        delta_counts.append(success)
        
        # Calculate confidence intervals
        if success > 0:  # Only calculate CI if there are successes
            se = np.sqrt(1/success + 1/(total-success) + 
                        1/(len(data[data['has_delta'] == 1]) - success) + 
                        1/(len(data[data['has_delta'] == 0]) - (total-success)))
            ci_lowers.append(np.exp(np.log(odds_ratio) - 1.96*se))
            ci_uppers.append(np.exp(np.log(odds_ratio) + 1.96*se))
        else:
            ci_lowers.append(0)
            ci_uppers.append(np.inf)
    
    results = pd.DataFrame({
        'Dimension': dimensions,
        'Odds_Ratio': odds_ratios,
        'P_Value': p_values,
        'CI_Lower': ci_lowers,
        'CI_Upper': ci_uppers,
        'Total_Count': total_counts,
        'Delta_Count': delta_counts
    })
    
    return results.sort_values('Odds_Ratio', ascending=False)

# Calculate odds ratios for both dimension types
regular_odds = calculate_odds_ratios_for_dimension(data, 'dimension')
semantic_odds = calculate_odds_ratios_for_dimension(data, 'semantic_dimension')

# Create visualization function
def plot_odds_ratios(odds_results, title):
    plt.figure(figsize=(12, 6))
    
    # Plot the odds ratios
    plt.scatter(range(len(odds_results)), odds_results['Odds_Ratio'], 
               color='blue', s=100, zorder=5)
    
    # Add error bars
    for i in range(len(odds_results)):
        if odds_results['Delta_Count'].iloc[i] > 0:  # Only plot CI if there are successes
            plt.vlines(i, 
                      odds_results['CI_Lower'].iloc[i],
                      odds_results['CI_Upper'].iloc[i],
                      color='blue', alpha=0.5)
    
    # Add horizontal line at OR = 1
    plt.axhline(y=1, color='r', linestyle='--', alpha=0.5, label='No Effect Line')
    
    # Customize plot
    plt.xticks(range(len(odds_results)), 
               [f"{dim}\n(n={count}, ={delta})" for dim, count, delta in 
                zip(odds_results['Dimension'], odds_results['Total_Count'], 
                    odds_results['Delta_Count'])],
               rotation=45,
               ha='right')
    plt.ylabel('Odds Ratio (log scale)')
    plt.xlabel('Dimension')
    plt.title(title)
    plt.yscale('log')
    
    # Add significance indicators
    for i, p in enumerate(odds_results['P_Value']):
        if odds_results['Delta_Count'].iloc[i] > 0:  # Only add stars if there are successes
            y_pos = odds_results['CI_Upper'].iloc[i]
            if p < 0.05:
                plt.text(i, y_pos*1.1, '*', ha='center', va='bottom', fontsize=12)
            if p < 0.01:
                plt.text(i, y_pos*1.1, '**', ha='center', va='bottom', fontsize=12)
            if p < 0.001:
                plt.text(i, y_pos*1.1, '***', ha='center', va='bottom', fontsize=12)
    
    plt.text(1.02, 0.98, '* p<0.05\n** p<0.01\n*** p<0.001', 
             transform=plt.gca().transAxes, 
             verticalalignment='top')
    
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

# Plot both sets of results
plot_odds_ratios(regular_odds, 'Odds Ratios by Original Dimension (with 95% CI)')
plot_odds_ratios(semantic_odds, 'Odds Ratios by Semantic Dimension (with 95% CI)')

# Print results
print("\nOriginal Dimension Odds Ratios:")
print(regular_odds[['Dimension', 'Odds_Ratio', 'P_Value', 'Total_Count', 'Delta_Count']].round(3))

print("\nSemantic Dimension Odds Ratios:")
print(semantic_odds[['Dimension', 'Odds_Ratio', 'P_Value', 'Total_Count', 'Delta_Count']].round(3))
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load your dataset
file_path = 'tendimension2_zeroshot_both.csv'
data = pd.read_csv(file_path)

# Prepare a contingency table: Calculate frequency counts of 'has_delta' and 'agree_label'
contingency_table = pd.crosstab(data['has_delta'], data['agree_label'])
print(f"Contingency Table:\n{contingency_table}")

# Perform Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)  # Perform the Chi-Square test

print("\nChi-Square Test Results:")
print(f"Chi2: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print(f"Expected frequencies:\n{expected}")

# Interpretation of the p-value
if p < 0.05:
    print("There is a significant relationship between the delta label and the agree label.")
else:
    print("There is no significant relationship between the delta label and the agree label.")
```
