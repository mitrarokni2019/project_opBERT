```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from transformers import RobertaTokenizer, RobertaModel
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

# Check GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Check GPU availability
print("Checking GPU availability...")
print(f"CUDA Available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU Device: {torch.cuda.get_device_name(0)}")
    print(f"Number of GPUs: {torch.cuda.device_count()}")
    # Set default tensor type to cuda
    torch.set_default_tensor_type('torch.cuda.FloatTensor')

# Install required packages if not already installed
!pip install -q transformers datasets evaluate

# Download the dataset using gdown
!pip install -q gdown
```

```python

# Load dataset
print("Loading dataset...")
#!gdown "https://drive.google.com/uc?id=10E9RV2p0XREskcaDeJMvQ430tgGDgq1Z"
#data = pd.read_csv('labeled_sample_data.csv')
# 3. Load and prepare data
#url for 6000= https://drive.google.com/file/d/1PS8MS_0DVu6Qj3eozRr8Hkmj-3KC4T23/view?usp=drive_link
!gdown "https://drive.google.com/uc?id=1PS8MS_0DVu6Qj3eozRr8Hkmj-3KC4T23"
print("\n=== Loading Dataset ===")
data = pd.read_csv('6000labeled_sample_data_all.csv', encoding='ISO-8859-1')

# Prepare data

# summarized_cleaned_bod
# summarized_cleaned_text
posts = data['summarized_cleaned_text'].tolist()
comments = data['summarized_cleaned_bod'].tolist()
#posts = data['cleaned_text'].tolist()
#comments = data['cleaned_body'].tolist()
labels = data['label_agreement'].map({-1: 0, 0: 1, 1: 2}).tolist()


print(f"Number of records in dataset: {len(comments)}")
print(f"Number of posts: {len(posts)}")
print(f"Number of comments: {len(comments)}")
print(f"Number of labels: {len(labels)}")

#distribution of label_agreement
print("Distribution of label_agreement:")
print(data['label_agreement'].value_counts())

#plotting distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='label_agreement', data=data)
plt.title('Distribution of label_agreement')
plt.xlabel('label_agreement')
plt.ylabel('Count')
plt.show()

# Split dataset
print("Splitting dataset...")
train_posts, val_posts, train_comments, val_comments, train_labels, val_labels = train_test_split(
    posts, comments, labels, test_size=0.2, random_state=42
)
```

```python
# Dataset class
class TextDataset(Dataset):
    def __init__(self, posts, comments, labels, tokenizer, max_length=512):
        self.combined_texts = [f"{post} </s> {comment}" for post, comment in zip(posts, comments)]
        self.labels = labels
        self.encodings = tokenizer(
            self.combined_texts,
            truncation=True,
            padding=True,
            max_length=max_length,
            return_tensors='pt'
        )

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        item = {
            'input_ids': self.encodings['input_ids'][idx],
            'attention_mask': self.encodings['attention_mask'][idx],
            'labels': torch.tensor(self.labels[idx])
        }
        return item
```

```python
# RoBERTa Model
class RobertaClassifier(nn.Module):
    def __init__(self, num_labels=3, dropout_rate=0.3):
        super().__init__()
        self.roberta = RobertaModel.from_pretrained('roberta-base')

        # Freeze some layers to prevent overfitting
        for param in self.roberta.embeddings.parameters():
            param.requires_grad = False

        self.dropout1 = nn.Dropout(dropout_rate)
        self.dropout2 = nn.Dropout(dropout_rate)

        # Additional layers
        self.dense = nn.Linear(self.roberta.config.hidden_size, 512)
        self.activation = nn.ReLU()
        self.layer_norm = nn.LayerNorm(512)
        self.classifier = nn.Linear(512, num_labels)

    def forward(self, input_ids, attention_mask, labels=None):
        outputs = self.roberta(input_ids=input_ids, attention_mask=attention_mask)

        # Get CLS token output
        pooled_output = outputs.last_hidden_state[:, 0, :]
        pooled_output = self.dropout1(pooled_output)

        # Dense layer with activation
        dense_output = self.dense(pooled_output)
        dense_output = self.activation(dense_output)
        dense_output = self.layer_norm(dense_output)
        dense_output = self.dropout2(dense_output)

        # Classification layer
        logits = self.classifier(dense_output)

        loss = None
        if labels is not None:
            loss_fct = nn.CrossEntropyLoss()
            loss = loss_fct(logits, labels)

        return {'loss': loss, 'logits': logits} if loss is not None else {'logits': logits}

# Initialize tokenizer
print("Initializing tokenizer...")
tokenizer = RobertaTokenizer.from_pretrained('roberta-base')

# Create datasets
print("Creating datasets...")
train_dataset = TextDataset(train_posts, train_comments, train_labels, tokenizer)
val_dataset = TextDataset(val_posts, val_comments, val_labels, tokenizer)
```

```python

# Create train and validation DataLoaders
train_loader = DataLoader(
    train_dataset,
    batch_size=16,
    shuffle=True,
    generator=torch.Generator(device=device)  # Ensure generator uses the same device
)

val_loader = DataLoader(
    val_dataset,
    batch_size=16,
    shuffle=False  # No shuffle for validation
)
```

```python
from transformers import AdamW

# Or use torch's AdamW:
from torch.optim import AdamW



# Initialize model
print("Initializing model...")
model = RobertaClassifier(dropout_rate=0.3).to(device)


# Freeze layers more systematically
for param in model.roberta.embeddings.parameters():
    param.requires_grad = False
for i, layer in enumerate(model.roberta.encoder.layer):
    if i < 6:  # Freeze first 6 layers
        for param in layer.parameters():
            param.requires_grad = False

# Optimizer with better parameters for Focal Loss
optimizer = AdamW(
    [p for p in model.parameters() if p.requires_grad],
    lr=2e-5,  # Lower learning rate for stability
    weight_decay=0.02,
    eps=1e-8
)


# Add learning rate scheduler
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer,
    mode='min',
    factor=0.5,
    patience=2,
    min_lr=1e-6,
    verbose=True
)



```

```python
import torch.nn.functional as F
from torch.utils.data import WeightedRandomSampler


# After dataset creation, calculate class weights
#labels_array = np.array(train_labels)
#class_counts = np.bincount(labels_array)
#total_samples = len(labels_array)
#class_weights = torch.FloatTensor([
#    1/count for count in class_counts
#]).to(device)
#print(f"Class weights: {class_weights}")




labels_array = np.array(train_labels)
class_counts = np.bincount(labels_array)
total_samples = len(labels_array)

raw_weights = [1/count for count in class_counts]
normalized_weights = [weight / sum(raw_weights) * len(class_counts) for weight in raw_weights]

class_weights = torch.FloatTensor(normalized_weights).to(device)
print(f"Class weights: {class_weights}")


# Calculate class weights (inverse of frequencies)
#class_weights = 1.0 / torch.FloatTensor(class_counts)
#sample_weights = [class_weights[label] for label in train_labels]
#sampler = WeightedRandomSampler(sample_weights, num_samples=len(sample_weights), replacement=True)



# Focal Loss implementation
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma

    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, weight=self.alpha, reduction='none')
        pt = torch.exp(-ce_loss)
        focal_loss = ((1 - pt) ** self.gamma) * ce_loss
        return focal_loss.mean()

# Initialize criterion with Focal Loss
criterion = FocalLoss(alpha=class_weights)

# Training settings
num_epochs = 5
best_val_f1 = 0
patience = 3
patience_counter = 0
metrics = {
    'train_loss': [], 'train_acc': [], 'train_f1': [],
    'val_loss': [], 'val_acc': [], 'val_f1': []
}

# Training loop
print("Starting training...")
for epoch in range(num_epochs):
    # Training phase
    model.train()
    train_loss = 0
    train_preds, train_true = [], []

    for batch in tqdm(train_loader):
        # Move batch to device
        batch = {k: v.to(device) for k, v in batch.items()}

        # Forward pass
        outputs = model(**batch)
        loss = criterion(outputs['logits'], batch['labels'])

        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Store predictions
        train_loss += loss.item()
        preds = torch.argmax(outputs['logits'], dim=1)
        train_preds.extend(preds.cpu().numpy())
        train_true.extend(batch['labels'].cpu().numpy())

    # Calculate training metrics
    avg_train_loss = train_loss / len(train_loader)
    train_acc = accuracy_score(train_true, train_preds)
    train_f1 = f1_score(train_true, train_preds, average='weighted')

    # Validation phase
    model.eval()
    val_loss = 0
    val_preds, val_true = [], []

    with torch.no_grad():
        for batch in tqdm(val_loader):
            batch = {k: v.to(device) for k, v in batch.items()}
            outputs = model(**batch)
            loss = criterion(outputs['logits'], batch['labels'])

            val_loss += loss.item()
            preds = torch.argmax(outputs['logits'], dim=1)
            val_preds.extend(preds.cpu().numpy())
            val_true.extend(batch['labels'].cpu().numpy())

    # Calculate validation metrics
    avg_val_loss = val_loss / len(val_loader)
    val_acc = accuracy_score(val_true, val_preds)
    val_f1 = f1_score(val_true, val_preds, average='weighted')

    # Store metrics
    metrics['train_loss'].append(avg_train_loss)
    metrics['train_acc'].append(train_acc)
    metrics['train_f1'].append(train_f1)
    metrics['val_loss'].append(avg_val_loss)
    metrics['val_acc'].append(val_acc)
    metrics['val_f1'].append(val_f1)

    # Early stopping check
    if val_f1 > best_val_f1:
        best_val_f1 = val_f1
        patience_counter = 0
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'val_f1': val_f1,
            'class_weights': class_weights
        }, 'best_model.pt')
        print(f"Saved new best model with F1: {val_f1:.4f}")
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print("Early stopping triggered!")
            break

    print(f"Epoch {epoch+1}")
    print(f"Train Loss: {avg_train_loss:.4f}, Train Acc: {train_acc:.4f}, Train F1: {train_f1:.4f}")
    print(f"Val Loss: {avg_val_loss:.4f}, Val Acc: {val_acc:.4f}, Val F1: {val_f1:.4f}")
```

```python
# Visualization of Loss and F1 Score
plt.figure(figsize=(15, 6))

# Loss visualization
plt.subplot(1, 2, 1)
plt.plot(metrics['train_loss'], label='Train Loss', marker='o')
plt.plot(metrics['val_loss'], label='Validation Loss', marker='o')
plt.title('Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid()

# F1 Score visualization
plt.subplot(1, 2, 2)
plt.plot(metrics['train_f1'], label='Train F1', marker='o')
plt.plot(metrics['val_f1'], label='Validation F1', marker='o')
plt.title('F1 Score Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('F1 Score')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()

# Visualization of Accuracy
plt.figure(figsize=(8, 6))
plt.plot(metrics['train_acc'], label='Train Accuracy', marker='o')
plt.plot(metrics['val_acc'], label='Validation Accuracy', marker='o')
plt.title('Accuracy Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid()
plt.show()

# Confusion Matrix Visualization
cm = confusion_matrix(val_true, val_preds)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Disagree', 'Neutral', 'Agree'],
            yticklabels=['Disagree', 'Neutral', 'Agree'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Classification Report
print("\nClassification Report:")
print(classification_report(val_true, val_preds,
                            target_names=['Disagree', 'Neutral', 'Agree'], digits=4))
```
